```python

embedding_size = 16
hidden_state_size = 32

graph = tf.Graph()
with graph.as_default(), tf.device('/cpu:0'):
    
    embeddings = tf.Variable(tf.random_uniform([vocabulary_size, embedding_size], -1.0, 1.0))
    
    w = tf.Variable(tf.truncated_normal([embedding_size + hidden_state_size, hidden_state_size], -0.1, 0.1))
    b = tf.Variable(tf.zeros([hidden_state_size]))
    
    
    w2 = tf.Variable(tf.truncated_normal([hidden_state_size, num_class], -0.1, 0.1))
    b2 = tf.Variable(tf.zeros([num_class]))
    
    inputs = tf.placeholder(tf.int32, shape=[window_size ,batch_size])
    labels = tf.placeholder(tf.int32, shape=[batch_size,num_class])
    
        
    def the_model(window_size, inputs):
        state = tf.Variable(tf.zeros([batch_size,hidden_state_size]), trainable=False)
        for i in range(0,window_size): 
            unrolled_inputs = inputs[i]
            embed = tf.nn.embedding_lookup(embeddings, unrolled_inputs)
            conc = tf.concat(1,[embed,state])
            state = tf.nn.relu(tf.matmul(conc, w)  +b)
        logits = tf.matmul(state, w2)  + b2
        state = tf.Print(state,[state],'state = ')
        return logits
    
    logits = the_model(window_size, inputs)
    loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits, labels))
    optimizer = tf.train.GradientDescentOptimizer(0.5).minimize(loss)
    
    train_prediction = tf.nn.softmax(logits)
    valid_prediction = tf.nn.softmax(the_model(window_size,zip(*tf_valid_inputs)))
    test_prediction = tf.nn.softmax(the_model(window_size,zip(*tf_test_inputs)))
    
```

```python
num_steps = 1600

            
def accuracy(predictions, labels):
  return (100.0 * np.sum(np.argmax(predictions, 1) == np.argmax(labels, 1))
          / predictions.shape[0])

with tf.Session(graph=graph) as session:
  tf.initialize_all_variables().run()
  print('Initialized')
  for step in range(num_steps):
    batch_train_inputs ,batch_train_labels= get_batch(batch_size,tf_train_inputs,tf_train_labels)
    feed_dict = {inputs: zip(*batch_train_inputs), labels: batch_train_labels}
    _, l, predictions = session.run([optimizer, loss, train_prediction],feed_dict = feed_dict)
    if (step % 100 == 0):
      print('Loss at step %d: %f' % (step, l))
#       print(b.eval())
#       print(predictions[0][:20])
#       print(state[0])
      print([max_index(i) for i in predictions[:20]])
#       print([max_index(i) for i in batch_train_labels])
      print('Training accuracy: %.1f%%' % accuracy(
        predictions, batch_train_labels))
      print('Validation accuracy: %.1f%%' % accuracy(
        valid_prediction.eval(), tf_valid_labels))
  print('Test accuracy: %.1f%%' % accuracy(test_prediction.eval(), tf_test_labels))
  print([max_index(i) for i in test_prediction.eval()])
  print([max_index(i) for i in tf_test_labels])



```
